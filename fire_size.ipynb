```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

MAIN_FOLDER = './figures/'

df = pd.read_csv('./data/FW_Veg_Rem_Combined.csv')

fire_size_classes = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

df_sorted = df.sort_values(by='fire_size_class')

sns.displot(data=df_sorted['fire_size_class'],  stat="probability")
plt.title(f'Distribution of fire_size_classes')
plt.savefig(f'{MAIN_FOLDER}fire_size_classes_distribution.png', bbox_inches='tight')
plt.show()
```
